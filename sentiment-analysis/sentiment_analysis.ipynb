```python
from textblob import TextBlob
def analyze_sentiment(text):
    blob = TextBlob(text)
    polarity = blob.sentiment.polarity
    if polarity > 0:
        sentiment = "Positive"
    elif polarity < 0:
        sentiment = "Negative"
    else:
        sentiment = "Neutral"
    return sentiment,polarity
text = input("Enter a sentence: ")
sentiment, polarity=analyze_sentiment(text)
print(sentiment)
print(polarity)
```
