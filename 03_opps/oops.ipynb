### OOPs concepts

```python
class Student:
    name="afnan"
s1=Student()
print(s1.name)   #object created with name s1

s2=Student()
print(s2.name)   #object created with name s2
```

```python
class car:
    color='red'
    brand='BMW'
car1=car()
print(car1.color) #give name 
print(car1)  #give object address
print(car1.brand)
```

##### init()


```python
class Student:
    name="afnan"
    def __init__(self): # self is an parameter 
        print("hello world")
s1=Student()
print(s1.name)
```

```python
class Student:
    def __init__(self, fullname):  #using multi-parameter 
        self.name=fullname
        print("hello world")
s1=Student('saif')
print(s1.name)
s2=Student('afnan')
print(s2.name)
print(s1.name+s2.name)
```

```python
class Student:
    def __init__(self, fullname,score):  #using multi-parameter 
        self.name=fullname
        self.mark=score
        print("add attribute")
s1=Student('saif',99)
print(s1.name,s1.mark)
s2=Student('afnan',90)
print(s2.name,s2.mark)
```

#### class & instance attributes

```python
class Student:
    college_name="adhiyamaan"
    def __init__(self, fullname,score):  #using multi-parameter 
        self.name=fullname
        self.mark=score
        print("add attribute")
s1=Student('saif',99)
print(s1.name,s1.mark)
s2=Student('afnan',90)
print(s2.name,s2.mark)
print(s2.college_name) #or
print(Student.college_name) 
```

```python
class Student:
    college_name="adhiyamaan"
    name="anonymous"  #class attr
    def __init__(self, fullname,score):  #using multi-parameter 
        self.name=fullname  #obj attr  > class attr
        self.mark=score
        print("add attribute")
s1=Student('saif',99)
print(s1.name,s1.mark)
print(s1.name)
```

#### methods

```python
class Student:
    college_name="adhiyamaan"
    def __init__(self, fullname,score):  #using multi-parameter 
        self.name=fullname  
        self.mark=score
    def welcome(self):
        print("welcome student,",self.name)
s1=Student('saif',99)
print(s1.name,s1.mark)
s1.welcome()
```

```python
class Student:
    def __init__(self,marks):
        self.marks=marks
    def get_marks(self):
        return self.marks
s1=Student(99)
print(s1.get_marks())
```

```python
class Student:
    def __init__(self, name, m1,m2,m3):
        self.name=name
        self.m1=m1
        self.m2=m2
        self.m3=m3
    def average(self):
        return (self.m1+self.m2+self.m3)/3
s1=Student("afnan",99,99,99)
print(s1.name,s1.m1,s1.m2,s1.m3)
print(s1.average())
```

```python
class Student:
    def __init__(self, name, marks):
        self.name=name
        self.marks=marks
    def average(self):
        sum=0
        for i in self.marks:
            sum+=i
        return sum/3
s1=Student("afnan",[99,99,99])
print(s1.name)
print(s1.average())
```
