```python
import numpy as np

scores = np.array([45, 78, 92, 55, 88, 63, 71, 84])

# 1. Average score
avg = np.mean(score)

# 2. Highest and lowest
high = np.max(score)
low = np.min(score)

# 3. Count above 70
above_70 = score[score>70]

# 4. Normalise to 0-100
normalised = (score - score.min()) / (score.max() - score.min()) * 100

print(f"Avg: {avg:.1f}, High: {high}, Low: {low}")
print(f"Above 70: {above_70}")
print(f"Normalised: {normalised}")
```

```python
import numpy as np

prices = np.array([100, 102, 98, 105, 110, 107, 115, 112, 118, 120])
window = 3

windows = np.lib.stride_tricks.sliding_window_view(prices, window)
rolling_mean = windows.mean(axis=1)
best_day = np.argmax(rolling_mean) + window - 1
overall_mean = rolling_mean.mean()
hot_days = np.where(rolling_mean > overall_mean)[0] + window - 1

print("Windows shape:", windows.shape)
print("Rolling mean:", rolling_mean.round(2))
print("Best 3-day window ends at index:", best_day)
print("Hot days (indices):", hot_days)
```
