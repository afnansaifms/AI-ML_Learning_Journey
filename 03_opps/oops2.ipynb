#### inheritance

```python
class Car:
    color="white"
    @staticmethod
    def start():
        print("..started")
    @staticmethod
    def stop():
        print("..stopped")
class Toyato(Car):
    def __init__(self, name):
        self.name=name
t1=Toyato("fortuner")
print(t1.name)
print(t1.color)
print(t1.start(),t1.stop())    
```

```python
class Car:
    @staticmethod
    def start():
        print("..started")
    @staticmethod
    def stop():
        print("..stopped")
        
class Toyato(Car):
    def __init__(self, brand):
        self.brand=brand
        
class Fortuner(Toyato):
    def __init__(self, model):
        self.model=model
t1=Fortuner("2021")
print(t1.start(),t1.stop())    
```

### super() method

```python
class Car:
    def __init__(self,type):
        self.type=type
    @staticmethod
    def start():
        print("..started")
    @staticmethod
    def stop():
        print("..stopped")
        
class Toyato(Car):
    def __init__(self, model,type):
        self.model=model
        super().start()
        super().__init__(type)
car1=Toyato("fortuner","hybrid")
print(car1.type)
    
```

```python
class Student:
    def __init__(self,m1,m2,m3):
        self.m1=m1
        self.m2=m2
        self.m3=m3
    @property
    def average(self):
        return str((self.m1+self.m2+self.m3)/3) +"%"
s1=Student(99,98,97)
print(s1.average)
s1.m3=100
print(s1.average)
        
```

### polymorphism

```python
class Complex:
    def __init__(self, real,img):
        self.real=real
        self.img=img
    def shownumber(self):
        print(self.real,"i +",self.img,"j")
    def add(self,num2):
        real=self.real+num2.real
        img=self.img+num2.img
        return Complex(real,img)

num1=Complex(2,3)
num2=Complex(3,2)
print(num1.shownumber())
print(num2.shownumber())
num3=num1.add(num2)
print(num3.shownumber())
```

```python
class Complex:
    def __init__(self, real,img):
        self.real=real
        self.img=img
    def shownumber(self):
        print(self.real,"i +",self.img,"j")
    def __add__(self,num2):           #usingdunder function
        real=self.real+num2.real
        img=self.img+num2.img
        return Complex(real,img)

num1=Complex(2,3)
num2=Complex(3,2)
print(num1.shownumber())
print(num2.shownumber())
num3=num1+num2
print(num3.shownumber())
```

```python
class Circle:
    def __init__(self,r):
        self.r=r
    def area(self):
        return 3.14*self.r*self.r
    def perimeter(self):
        return 2*3.14*self.r

c=Circle(5)
print(c.area())
print(c.perimeter())
```
